```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Task 1: Collect historical closing prices from Yahoo Finance for GOOG, AMZN, and SPY from September 1st, 2018, to September 1st, 2023.

```python
symbols = ["GOOG", "AMZN", "SPY"]

start_date = "2018-09-01"
end_date = "2023-09-01"

monthly_closing_prices = pd.DataFrame()

for symbol in symbols:
    stock_data = yf.download(symbol, start=start_date, end=end_date)
    monthly_closing_prices[symbol] = stock_data["Adj Close"].resample('M').ffill()
```

# Calculate the percentage change in closing prices for each stock.

```python
monthly_returns = monthly_closing_prices.pct_change()
monthly_returns = monthly_returns.dropna()
```

```python
x = monthly_returns['SPY'].values
y = monthly_returns['GOOG'].values
z = monthly_returns['AMZN'].values
```

```python
plt.scatter(x,y)
plt.title('Scatter plot: GOOG vs SPY')
plt.show()
```

```python
plt.scatter(x,z)
plt.title('Scatter plot: AMZN vs SPY')
plt.show()
```

# Task 3: Implement machine learning techniques to determine α and β using different learning rates and numbers of epochs.

```python
epochs = 5000
lr = 0.001
m = len(x)
x = monthly_returns['SPY'].values
y = monthly_returns['GOOG'].values
b0 = 1
b1 = 1.2

error_list =[]

def gradient(b0,b1):

    db0 = -2*(np.sum(y- b0 - b1*x))/m
    db1 = -2*(np.dot(x ,(y- b0 - b1*x)))/m

    b0 -= db0 * lr
    b1 -= db1*lr
    return b0, b1

for i in range(epochs):
    error = np.sum(np.square(y-b0 - b1*x))/m
    error_list.append(error)
    b0 , b1 = gradient(b0,b1)

print("Intercept=", b0, "Slope=", b1, "error =", error)

plt.plot(error_list)
```
